```python
import torch

# load RoBERTa model (takes a long time to run this code block)
roberta = torch.hub.load('pytorch/fairseq', 'roberta.large')
roberta.eval() # change the model to eval mode
```

```python
# test byte pair encoding
tokens = roberta.encode('Hello world!')
assert tokens.tolist() == [0, 31414, 232, 328, 2]
roberta.decode(tokens)  # should be 'Hello world!'
```
